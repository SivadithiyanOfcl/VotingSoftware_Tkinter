```python
from tkinter import *
import sqlite3 as sq3
from tkinter import messagebox

from login import *
from voter import *
from register import *
from operator_file import *



#Creating the main database
conn=sq3.connect('main.db',timeout = 15) #main database
cursor=conn.cursor() #main cursor

#Creating global variables

global currentpoll

cursor.execute("""CREATE TABLE IF NOT EXISTS Buffer (Current_poll TEXT, active TEXT, Winner TEXT)""")
conn.commit()
cursor.execute("SELECT Current_poll FROM Buffer")
buf = cursor.fetchall()

if buf == [] or buf == '':
    cursor.execute("INSERT INTO Buffer VALUES (?,?,?)",("Default","yes","Default Winner"))
    conn.commit()
else:
    cursor.execute("SELECT Current_poll from Buffer WHERE active = 'yes'")
    conn.commit()
    cpoll = cursor.fetchall()
    cpoll = cpoll[0][0]
    
    if cpoll == "Default":
        currentpoll = cpoll
        #messagebox.showinfo("Info","Runnning in default mode")
    else:
        currentpoll = cpoll

#Creating the main page
root = Tk()
root.title("Voting")
root.geometry("700x500")
root.resizable(False,False)



#Declaring variables to get & store the input (username,password)

u_name = StringVar()
p_word = StringVar()  

def login_set():
    
    global currentpoll
    
    us = usernameEntry.get()
    pa = passwordEntry.get()
    login(us,pa,currentpoll)
    
def vo_main():
    vote_main(currentpoll,u_name)
    
def op_main():
    operator_main()
    
def exit():
    root.destroy()
 
    #bg.png
    
backGroundImage = PhotoImage(file ="bg.png")
backGroundImageLabel = Label(root,image = backGroundImage)
backGroundImageLabel.place(x=0,y=0)

loginButton = Button(root,text ="login",command=login_set)
loginButton.place(x=275,y=470)

registerButton = Button(root,text = "register",command = register)
registerButton.place(x=400,y=470) 

canvasLabel = Canvas(root,width = 400,height = 100)
canvasLabel.place(x=150,y = 360)  

titleLabel = Label(root,text = "Login",font = "Bold 30")
titleLabel.place(x=300,y=20)     

userNameLabel = Label(root,text = "Username",font = "8")
userNameLabel.place(x=200,y=380)  

passwordLabel = Label(root,text = "Password",font = "8")
passwordLabel.place(x=200,y=420)  

usernameEntry = Entry(root,borderwidth = 0,highlightthickness = 0,width = 15,textvariable=u_name)
usernameEntry.place(x=400,y=380)

passwordEntry = Entry(root,borderwidth = 0,show = "*",highlightthickness = 0,textvariable=p_word)
passwordEntry.place(x=400,y=420,width=100,height=20)

exitButton = Button(root,text = "Exit",command =exit)
exitButton.place(x=650,y=470)

buttonn = Button(root,text = "Op",command =op_main)
buttonn.place(x = 10, y =10)

buttonn = Button(root,text = "Voter",command =vo_main)
buttonn.place(x = 50, y =10)


#---------------------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------------

#Main loop

if __name__ == "__main__":

    #create a database table
    conn=sq3.connect('main.db',timeout = 15) #main database
    cursor=conn.cursor() #main cursor

    root.mainloop()

#---------------------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------------

```
